#  Set-up DataView using EM analytic functions

```python
x = np.linspace(-80., 80., 200)
y = np.linspace(-80., 80., 200)
self = DataView()
srcLoc = np.r_[0., 0., 0.] # source location
sig, f = 1e0, np.r_[1e6]  # conductivity (S/m), frequency (Hz)
self.set_xyz(np.r_[0.],x,y, normal="X") # set plane and locations ... 
self.eval_2D(srcLoc, sig, f, "x", EM.Analytics.E_from_ElectricDipoleWholeSpace) # evaluate
```
